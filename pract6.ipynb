```python
import numpy as np #импортириует библиотеку numpy
import pandas as dp #импортириует библиотеку pandas
import matplotlib.pyplot as plt #импортириует библиотеку matplotlib.pyplot (для того чтобы делать графики)
df = dp.read_csv('waterQuality1.csv') #читает таблицу/файл
x = df["arsenic"].head(5) #выводит значения из пяти строчек столбца arsenic 
y = df["aluminium"].head(5) #выводит значения из пяти строчек столбца aluminium

plt.plot(x, y, label='Линейный график') #используется функция для создания линейного графика, label нужен для legend, чтобы отображалось та надпись, которую принимает переменная label
plt.xlabel('Ось x') #обозначение оси x
plt.ylabel('Ось y') #обозначение оси y
plt.grid() #сетка на графика
plt.legend() #Легенда с описаниями элементов графика.
plt.show() #Отображение текущего графика.
#Вывод: сделан график, отображающий значения из таблицы (arsenic, aluminium) в виде линейного графика
```

```python
import numpy as np #импортириует библиотеку numpy
import pandas as dp #импортириует библиотеку pandas
import matplotlib.pyplot as plt #импортириует библиотеку matplotlib.pyplot (для того чтобы делать графики)
df = dp.read_csv('waterQuality1.csv') #читает таблицу/файл

x = ['aluminium', 'ammonia', 'arsenic', 'barium', 'cadmium'] #значения из таблицы, которые принимает переменная x
y = [float(df["aluminium"].head(1)), float(df["ammonia"].head(1)), float(df["arsenic"].head(1)), float(df["barium"].head(1)), float(df["cadmium"].head(1))] #значения из таблицы, которые принимает переменная y

plt.bar(x, y, label='Величина элементов', color='red') #делает таблицу, цвет для неё и надпись
plt.xlabel('Элемент') #надпись для значений x
plt.ylabel('Значение (Величина)') #надпись для значений y
plt.title('Столбчатая диаграмма элементов') #название для графика
plt.legend() #подписывает какой цвет отвечает за что (в строке с функцией, которая отвечает за сам график, есть label, он отвечает за эту надпись)
plt.show() #выводит график
#Вывод: сделан график, сравнивающий значения элементов из таблицы ('aluminium', 'ammonia', 'arsenic', 'barium', 'cadmium') в виде столбчатой диаграммы
```

```python
import numpy as np #импортириует библиотеку numpy
import pandas as dp #импортириует библиотеку pandas
import matplotlib.pyplot as plt #импортириует библиотеку matplotlib.pyplot (для того чтобы делать графики)
df = dp.read_csv('waterQuality1.csv') #читает таблицу/файл

numb = [float(df["aluminium"].head(1)), float(df["ammonia"].head(1)), float(df["arsenic"].head(1)), float(df["barium"].head(1)), float(df["cadmium"].head(1))] #принимает значения элементов из таблицы
ele = ["aluminium", "ammonia", "arsenic", "barium", "cadmium"] #принимает названия элемнтов из таблицы

plt.pie(numb, labels=ele, autopct='%1.1f%%') #делает круговую диаграмму
plt.title("Сравнение значений элементов") #присваивает название диаграмме
plt.show #показывает диаграмму
#Вывод: сделан график, которая сравнивает значения элементов в виде круговой диаграммы
```

```python
import numpy as np #импортириует библиотеку numpy
import pandas as dp #импортириует библиотеку pandas
import matplotlib.pyplot as plt #импортириует библиотеку matplotlib.pyplot (для того чтобы делать графики)
df = dp.read_csv('waterQuality1.csv') #читает таблицу/файл

x = [df["chloramine"].head(100)] #значения из таблицы, который принимает x
y = [df["chromium"].head(100)] #значения из таблицы, который принимает y
plt.figure(figsize=(10, 6)) #параметр для задания размера фигуры (ширина, высота) в дюймах.
plt.scatter(x, y) #делает таблицу рассеивания по переменным x и y
plt.title('Связи между элементами') #делает заголовок
plt.xlabel('Элемент "chloramine"') #Подпись по оси X текущего графика.
plt.ylabel('Элемент "chromium') #Подпись по оси Y текущего графика.
plt.show() #выводит график
#Вывод: сделан график в виде диаграммы рассеивания, который сравнивает значения между собой
```
